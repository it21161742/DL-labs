```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Create a random 4x4 array from an exponential distribution

```python
exponential_array = np.random.exponential(scale=1.0, size=(4, 4))
print("4x4 Exponential Distribution Array:")
print(exponential_array)
```

# Create a random 100000x1 array from the exponential distribution

```python
exponential_data = np.random.exponential(scale=1.0, size=100000)
print(exponential_data)
```

# Generate data for uniform and normal distributions

```python
uniform_data = np.random.rand(100000)
normal_data = np.random.randn(100000)
print(uniform_data)
print(normal_data)
```

# Plot histograms of the distributions

```python
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(normal_data, density=True, bins=100, histtype="step", color="green", label="Normal")
plt.hist(exponential_data, density=True, bins=100, histtype="step", color="red", label="Exponential")

plt.axis([-1, 8, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

3 . from mpl_toolkits.mplot3d import Axes3D

```python
# Generate X and Y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Compute Z
z = x**2 + y**2

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z, cmap='viridis')

# Add labels
ax.set_title('3D plot of Z = X^2 + Y^2')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

4 . Calculate the pearson standard coefficient and spearman rank correlation for the [HP,
Attack, Defense, Sp. Ark, Sp. Def, Speed] features in the seaborn tutorial.ipynb. Visualize the
results using heatmaps. Make sure the values are displaced in the heatmap as well.


```python
import seaborn as sns
from scipy.stats import spearmanr
```

```python
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
```

```python
# Display first 5 observations
df.head()
```

```python
# Pearson correlation
features = ["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]
pearson_corr = df[features].corr(method='pearson')
plt.figure(figsize=(10, 8))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Pearson Correlation Coefficient')
plt.show()

```

```python
# Spearman correlation
spearman_corr = df[features].corr(method='spearman')
plt.figure(figsize=(10, 8))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Spearman Rank Correlation')
plt.show()
```
